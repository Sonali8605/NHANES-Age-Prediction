```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv("NHANES_age_prediction.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df["age_group"].value_counts()
```

```python
X=df.drop(columns=["age_group"])
X
```

```python
y=df["age_group"]
y
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
len(X_train),len(X_test),len(y_train),len(y_test)
```

```python
scaler=StandardScaler()
scaler_X_train=scaler.fit_transform(X_train)
scaler_X_test=scaler.transform(X_test)
```

```python
scaler_X_train.std()
```

```python
model=LogisticRegression()
```

```python
model.fit(scaler_X_train,y_train)
```

```python
y_pred=model.predict(scaler_X_test)
y_test,y_pred
```

```python
model.coef_
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,classification_report,confusion_matrix
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
precision=precision_score(y_test,y_pred, pos_label="Adult")
precision
```

```python
precision1=precision_score(y_test,y_pred, pos_label="Senior")
precision
```

```python
recall=recall_score(y_test,y_pred,pos_label="Adult")
recall
```

```python
recall_1=recall_score(y_test,y_pred,pos_label="Senior")
recall_1
```

```python
c_m=confusion_matrix(y_test,y_pred)
c_m
```

```python
clr=classification_report(y_test,y_pred)
print(clr)
```

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr,_ = roc_curve(y_test,y_pred)
plt.figure(figsize=(10,8))
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()
```
